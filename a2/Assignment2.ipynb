Answer the following questions with mathematical analysis:

# Problem 4 (15% for CSC 446, 10% for CSC 546)

Let $W$ and $W_Q$ denote the total waiting time of a packet in the system and the queueing time of the packet, respectively. 

1. Is $W$ the same in an M/M/1 model having arrivals at rate λ and service at rate 2µ as it is in a two-server M/M/2 model with arrivals at rate λ and with each server at rate µ? Prove your claim. 

$\rho = \frac{\lambda}{2\mu}$

$P_0 = [1 + 2\rho + \frac{2\rho^2}{1-\rho}]^{-1} = [\frac{1 + \rho}{1 - \rho}]^{-1} = \frac{1 - \rho}{1 + \rho}$

$L = 2\rho + \frac{(2\rho)^{3}(1-\rho)}{4(1-\rho)^{2}(1+\rho)} = \frac{2\rho^3}{(1-\rho^2)}$

$L = \frac{4\mu\lambda}{(2\mu-\lambda)(2\mu+\lambda)}$

$W_2 = \frac{L}{\lambda} = \frac{4\mu}{(2\mu-\lambda)(2\mu+\lambda)}$

Since $W_1 = \frac{1}{2\mu-\lambda} \ne W_2$ 

Ans: $W(M/M/2) \ne W(M/M/1)$

2. Is $W_Q$ the same in an M/M/1 model having arrivals at rate λ and service at rate 2µ as it is in a two-server M/M/2 model with arrivals at rate λ and with each server at rate µ? Prove your claim.

For $W_Q$ in M/M/1

$W_Q = \frac{\lambda}{2\mu(2\mu-\lambda)}$

For $W_Q$ in M/M/2

$W_Q = W - \frac{1}{\mu} = \frac{\lambda^2}{\mu(2\mu-\lambda)(2\mu+\lambda)}$

Compare $\frac{\lambda}{2\mu(2\mu-\lambda)}$ with $\frac{\lambda^2}{\mu(2\mu-\lambda)(2\mu+\lambda)}$

And compare $\frac{1}{2}$ with $\frac{\lambda}{2\mu+\lambda}$

$\rho = \frac{\lambda}{2\mu} \lt 1$ 

Therefore $\frac{1}{2} \gt \frac{\lambda}{2\mu+\lambda}$ and $\frac{\lambda}{2\mu(2\mu-\lambda)} \gt \frac{\lambda^2}{\mu(2\mu-\lambda)(2\mu+\lambda)}$

Ans: $W_Q(M/M/1) \ne W_Q(M/M/2)$
